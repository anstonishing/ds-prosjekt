### Step 2.1 - Clean data

```python
from sklearn.model_selection import train_test_split
# Stratified sampling to reduce dataset size while maintaining target representation (OpenAI, personal communication, February 24, 2023)
_, reduced_df = train_test_split(df, train_size=51337, stratify=df['Target'], random_state=42)

# Reset index (So, 2023)
reduced_df.reset_index(drop=True, inplace=True)

# Display the new shape of the dataset and target variable distribution
print("Reduced dataset shape:", reduced_df.shape)
print(reduced_df['Target'].value_counts())
```

```python
# Create a copy of the dataframe
df_cleaned = reduced_df.copy()
# (So, 2023)
```

```python
# Unit Tests to see if the original df was copied correctly
assert isinstance(df_cleaned, pd.DataFrame)
assert df_cleaned.shape == reduced_df.shape
# (So, 2023)
```

```python
# Drop columns
df_cleaned = df_cleaned.drop(columns=['age_band', 'gender', 'ID'])
# (So, 2023)
```

### 2.2 Select data by feature engineering

- Total warranty services could indicate how much the customer values maintenance and repairs, which could be a good indicator of their likelihood to purchase a new car.

- Average number of services per year could give an idea of how frequently the customer uses their car and how well they maintain it, which could also be useful indicators of their likelihood to purchase a new car.

```python
# Total warranty services
df_cleaned['total_warranty_services'] = df_cleaned['sched_serv_warr'] + df_cleaned['non_sched_serv_warr']

# Average number of services per year
df_cleaned['avg_num_services_per_year'] = df_cleaned['total_services'] / df_cleaned['age_of_vehicle_years']
```

```python
# Check if the new features has a strong relationship with the target value
from scipy.stats import pointbiserialr

target = df_cleaned['Target']
total_warranty_services = df_cleaned['total_warranty_services']
avg_num_services_per_year = df_cleaned['avg_num_services_per_year']

correlation_total_warranty_services = pointbiserialr(target, total_warranty_services)
correlation_avg_num_services_per_year = pointbiserialr(target, avg_num_services_per_year)

print("Correlation between total_warranty_services and target:", correlation_total_warranty_services)
print("Correlation between avg_num_services_per_year and target:", correlation_avg_num_services_per_year)

# (SciPy, 2014)
```

```python
# Drop avg_num_services_per_year because it has a very weak relationship
df_cleaned = df_cleaned.drop(columns=['avg_num_services_per_year'])
```

### 3.1 Generate test design by splitting data

ID column is not included to avoid overfitting as it is used to uniquely identify each record in the dataset and does not contribute to the prediction task.

Write why 75:25 ratio was chosen - Default value in sklearn

```python
# Extract features and target variable into Y and X
y = df_cleaned["Target"]
x = df_cleaned.drop("Target", axis=1)

# Import train_test_split from sklearn.model_selection - Already done in Step 2 but used to compile crucial info into Step 3
from sklearn.model_selection import train_test_split

# Split data into training and testing sets with 75:25 ratio
x_data, x_test, y_data, y_test = train_test_split(x, y, test_size=0.25, random_state=69)

# Split data into training and validation sets with 75:25 ratio 
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.25, random_state=69)

# (So, 2023)
```

### 3.2 Import relevant libraries, prepare features for modelling

```python
from sklearn.compose import ColumnTransformer # Used to transform columns in DF for specified rules
from sklearn.preprocessing import OneHotEncoder # Used to convert categorical variables into binary features
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, make_scorer # Performance metrics for the modelling

# Define the categorical features - Updated because some were removed in Step 2.1
categorical_features = ['car_model', 'car_segment']

# Create the ColumnTransformer - (OpenAI, personal communication, February 24, 2023) 
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)],
    remainder='passthrough'
)
```

### 3.3 Create baseline model

```python
# Find the mode of Y
y_mode = y.mode()

# Create a baseline prediction array with the mode of Y
baseline_prediction = np.full_like(y_train, y_mode)

# Evaluate the performance of the baseline model
accuracy = accuracy_score(y_train, baseline_prediction)

print("Baseline Model Accuracy:", accuracy)

# (So, 2023)
```

###  3.4 - Experiment 1: Train Logistic Regression Classifier

```python
from sklearn.linear_model import LogisticRegression

# Preprocess the training and validation sets
X_train_preprocessed = preprocessor.fit_transform(x_train)
X_val_preprocessed = preprocessor.transform(x_val)

# Instantiate the Logistic Regression model - (OpenAI, personal communication, February 24, 2023) - Prompt: Fix warning regarding failed convergence
log_reg = LogisticRegression(max_iter=2000)

# Fit the model with the training data
log_reg.fit(X_train_preprocessed, y_train)

# Make predictions on the training set
y_preds_train = log_reg.predict(X_train_preprocessed)

# Calculate the metrics for the training set
train_accuracy = accuracy_score(y_train, y_preds_train)
train_precision = precision_score(y_train, y_preds_train)
train_recall = recall_score(y_train, y_preds_train)
train_f1 = f1_score(y_train, y_preds_train)

# Make predictions on the validation set
y_preds_val = log_reg.predict(X_val_preprocessed)

# Calculate the metrics for the validation set
val_accuracy = accuracy_score(y_val, y_preds_val)
val_precision = precision_score(y_val, y_preds_val)
val_recall = recall_score(y_val, y_preds_val)
val_f1 = f1_score(y_val, y_preds_val)

# Display the results - (OpenAI, personal communication, February 24, 2023) Used to present results in a structured way
print("Training Set Metrics:")
print(f"Accuracy: {train_accuracy}")
print(f"Precision: {train_precision}")
print(f"Recall: {train_recall}")
print(f"F1: {train_f1}")

print("\nValidation Set Metrics:")
print(f"Accuracy: {val_accuracy}")
print(f"Precision: {val_precision}")
print(f"Recall: {val_recall}")
print(f"F1: {val_f1}")

# (So, 2023)
```

```python
from sklearn.model_selection import GridSearchCV

# Define hyperparameter grid
param_grid = {
    'C': [0.1, 1, 10],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear', 'saga']
}

# Instantiate the logistic regression model
log_reg = LogisticRegression(max_iter=1000)

# Create the GridSearchCV object 
grid_search = GridSearchCV(log_reg, param_grid, scoring='f1', verbose=1, n_jobs=-1)

# Fit the GridSearchCV object to the preprocessed data
grid_search.fit(X_train_preprocessed, y_train)

# Get the best parameters
best_params = grid_search.best_params_
best_estimator = grid_search.best_estimator_

print("Best parameters found:", best_params)

# Make predictions on the validation set using the best estimator
y_preds_val_best = best_estimator.predict(X_val_preprocessed)

# Calculate the metrics for the validation set using the best estimator
val_accuracy_best = accuracy_score(y_val, y_preds_val_best)
val_precision_best = precision_score(y_val, y_preds_val_best)
val_recall_best = recall_score(y_val, y_preds_val_best)
val_f1_best = f1_score(y_val, y_preds_val_best)

print("\nValidation set metrics (best estimator):")
print("Accuracy:", val_accuracy_best)
print("Precision:", val_precision_best)
print("Recall:", val_recall_best)
print("F1-score:", val_f1_best)

# (So, 2023), (OpenAI, personal communication, February 24, 2023) Used to debug issues creating GridSearchCV object
```

```python
from sklearn.model_selection import StratifiedKFold, cross_validate
from sklearn.metrics import make_scorer

# Preprocess the entire feature set 'x'
x_preprocessed = preprocessor.fit_transform(x)

# Define the stratified k-fold cross-validation
stratified_k_fold = StratifiedKFold(n_splits=5, random_state=42, shuffle=True)

# Define the scoring metrics
scoring_metrics = {
    'accuracy': 'accuracy',
    'precision': make_scorer(precision_score),
    'recall': make_scorer(recall_score),
    'f1': make_scorer(f1_score)
}

# Instantiate the Logistic Regression model with the best hyperparameters
log_reg_best = LogisticRegression(C=10, penalty='l2', solver='liblinear', max_iter=1000)

# Perform the cross-validation and calculate the metrics
cv_results_best = cross_validate(log_reg_best, x_preprocessed, y, cv=stratified_k_fold, scoring=scoring_metrics)

# Display the cross-validation results
print("Cross-Validation Metrics with Best Hyperparameters:")
print("Mean Accuracy:", cv_results_best['test_accuracy'].mean())
print("Mean Precision:", cv_results_best['test_precision'].mean())
print("Mean Recall:", cv_results_best['test_recall'].mean())
print("Mean F1 Score:", cv_results_best['test_f1'].mean())

# (So, 2023), (Sklearn.Model_Selection.StratifiedKFold, n.d.), (Explain Stratified K Fold Cross Validation in ML in Python, 2022)
```

```python
from sklearn.model_selection import StratifiedKFold, cross_validate
from sklearn.pipeline import Pipeline

# Create the pipeline for the preprocessing and the SVC model
svc_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', SVC(kernel='rbf', class_weight='balanced'))
])

# Instantiate the StratifiedKFold object
skf = StratifiedKFold(n_splits=5)

# Define scoring metrics
scoring = {'accuracy': make_scorer(accuracy_score),
           'precision': make_scorer(precision_score, zero_division=0),
           'recall': make_scorer(recall_score),
           'f1': make_scorer(f1_score)}

# Perform stratified cross-validation
cv_results = cross_validate(svc_pipeline, x, y, cv=skf, scoring=scoring, n_jobs=-1)

# Print the results
print("Stratified Cross-Validation Metrics:")
print(f"Accuracy: {np.mean(cv_results['test_accuracy'])}")
print(f"Precision: {np.mean(cv_results['test_precision'])}")
print(f"Recall: {np.mean(cv_results['test_recall'])}")
print(f"F1: {np.mean(cv_results['test_f1'])}")

# (So, 2023), (Sklearn.Model_Selection.StratifiedKFold, n.d.), (Explain Stratified K Fold Cross Validation in ML in Python, 2022)
```
