### Step 1.4 - Explore data

```python
import pandas as pd
from scipy.stats import pearsonr, pointbiserialr, f_oneway

# Read the data
df = pd.read_csv("loan_data_training.csv")

# Extract the target variable
target = df['TARGET']

# Create lists to store feature types
numeric_features = []
binary_categorical_features = []
non_binary_categorical_features = []

# Determine the type of each feature
for column in df.columns:
    if df[column].dtype in ['int64', 'float64'] and column != 'TARGET':
        numeric_features.append(column)
    elif df[column].dtype == 'object':
        unique_values = df[column].nunique()
        if unique_values == 2:
            binary_categorical_features.append(column)
        elif unique_values > 2:
            non_binary_categorical_features.append(column)

# Compute Pearson correlation for numeric features
numeric_correlations = {}
for feature in numeric_features:
    if df[feature].notna().all(): # Only use if there are no missing values
        numeric_correlations[feature] = pearsonr(df[feature], target)[0]

# Compute Point-Biserial correlation for binary categorical features
binary_correlations = {}
for feature in binary_categorical_features:
    if df[feature].notna().all(): # Only use if there are no missing values
        df[feature] = pd.Categorical(df[feature])
        binary_correlations[feature] = pointbiserialr(df[feature].cat.codes, target)[0]

# Compute ANOVA for non-binary categorical features
anova_results = {}
for feature in non_binary_categorical_features:
    categories = list(df[feature].dropna().unique())
    data = [target[df[feature] == category] for category in categories]
    F, p = f_oneway(*data)
    anova_results[feature] = {'F-value': F, 'p-value': p}

print("Numeric Feature Correlations:")
print(numeric_correlations)
print("\nBinary Categorical Feature Correlations:")
print(binary_correlations)
print("\nANOVA results for Non-Binary Categorical Features:")
print(anova_results)

# (Brownlee, 2020)
# (DeJesus, 2021)
# (Dong, 2023)
# (Lee, 2022)
# (So, 2023)
# (OpenAI, 2023)
```
